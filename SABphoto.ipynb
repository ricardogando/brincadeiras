Quantas pessoas na foto da SAB?

```python
import numpy as np
from skimage import data, io, filters
from PIL import Image
%matplotlib inline
```

Lendo a foto da SAB e convertendo num array 

```python
jpg = Image.open('sab2017.jpeg')
bbox = list(jpg.getbbox())
img = np.array(list(jpg.getdata())).T
```

Pegando apenas um canal

```python
img0 = img[0]
img0.shape = (bbox[3], bbox[2])
img0 = img0.astype(np.int8) # filter works with int8 only?
```

```python
io.imshow(img0)
```

Procurando as bordas

```python
edg = filters.sobel(img0)

io.imshow(edg)
```
